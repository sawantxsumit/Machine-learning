```python
import pandas as pd
import seaborn as sns
```

```python
sales=pd.read_excel("sales.xlsx")
```

```python
# sns.set_theme(style="whitegrid", palette="pastel")
```

```python
#categorical scatter plot
sns.catplot(x='Ship Mode', y='Sales' , data=sales, hue='Ship Mode');
```

```python
sns.catplot(x='Ship Mode', y='Sales' , data=sales, hue='Ship Mode', jitter= False);
#jitter is used when we dont want to see the magnitude of the data
```

```python
sns.catplot(x='Ship Mode', y='Sales' , data=sales.query("Profit > 1000"), hue='Ship Mode');
```

```python
sns.catplot(x='Ship Mode', y='Sales' , data=sales.query("Profit > 4000"), hue='Ship Mode',kind='swarm' );
```

```python
sns.catplot(y='Ship Mode', x='Sales' , data=sales.query("Profit > 4000"), hue='Ship Mode',kind='swarm' );
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', data= sales , kind='box', hue='Product Category');
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', data= sales , kind='box', hue='Ship Mode');
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', data= sales , kind='box', row='Ship Mode');
```

```python
sns.catplot(x='Product Category', y='Shipping Cost', data= sales , kind='box',col='Ship Mode',row='Customer Segment', hue='Ship Mode');
```

```python
#violin chart
sns.catplot(x='Product Category' , y='Shipping Cost' , data= sales ,hue='Product Category', kind='violin');
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data= sales ,hue='Ship Mode', kind='violin');
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data= sales ,hue='Ship Mode', inner= 'stick',kind='violin');
```

```python
#bar chart
sns.catplot(x='Product Category' , y='Sales' , data= sales ,col='Customer Segment', col_wrap=2,hue='Ship Mode', kind='bar');
```

```python
sns.catplot(y='Product Category' , x='Sales' , data= sales ,col='Customer Segment', col_wrap=2,hue='Ship Mode', kind='bar');
```

```python
#point plot
sns.catplot(x='Product Category' , y='Sales', data= sales, kind='point', errorbar=None);
```

```python
sns.catplot(x='Product Category' ,col='Ship Mode' , col_wrap=2, y='Sales', data= sales,hue='Ship Mode', kind='point', errorbar=None);
```

```python
sns.catplot(x='Product Category' , y='Sales',col='Customer Segment',col_wrap=2, data= sales, kind='point',errorbar=None, hue='Region');
```
