```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
#tips and penguins are example dataset already available in seaborn which we can use for practice
#we cannot load custom dataset using this function
```

```python
df.head()
```

```python
df.dtypes
```

corelation with heatmap

```python
cor_df=df.select_dtypes(include=['number']).corr()
```

```python
sns.heatmap(cor_df)
```

JOINPLOT
a plot that allows to study the realtionship bw 2 numeric values. the central chart display their corealtion. 
it is usually a scatterplot , a hex plot, a 2d histogram , or a density plot

UNIVARIATE ANALYSIS (analysing one feature variable only)

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

PAIR PLOT
in this plot one variable is matched with  another variable's value . 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df , hue='sex')
```

```python
sns.pairplot(df , hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
sns.distplot(df['tip'], kde=False)
```

```python
sns.displot(df['tip'],bins=20, kde=True)
```

```python
#countplot
sns.countplot(x='sex', data=df, hue='sex')
```

```python
sns.countplot(x='smoker', data=df, hue='sex')
```

```python
sns.countplot(y='day', data=df, hue='sex')
```

```python
#bar plot
sns.barplot(y='total_bill', x='sex', data=df, hue='sex', errorbar=None)
```

```python
#box plot
sns.boxplot(x='sex', y='total_bill' , data=df)
```

```python
sns.boxplot(x='day', y='total_bill' , data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
#violin plot
sns.violinplot(x='day' , y='total_bill' , data=df, hue='day' )
```
