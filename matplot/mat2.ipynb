```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,100)
y1= np.sin(x)
y2= np.cos(x)
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(x, y1, label="sin(x)" , color= 'red');
ax1.set_title("Sine Wave");
ax1.legend();
```
